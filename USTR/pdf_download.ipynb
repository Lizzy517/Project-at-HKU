```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
import os
import requests
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

```

```python

# 使用ChromeDriverManager来自动获取Chrome WebDriver
driver = webdriver.Chrome()

# 打开网页
url = 'https://comments.ustr.gov/s/docket?docketNumber=USTR-2023-0001'
driver.get(url)

while True:
    # 等待一段时间
    time.sleep(5)  
    # 获取页面源代码
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')

    # 查找表体中的每一行
    # 查找表体中的每一行
    tbody = soup.find('table').find('tbody') if soup.find('table') else None
    if tbody:
    # 遍历每一行，提取评论信息
        for row in tbody.find_all('tr'):
            # 提取 Submission ID
            submission_id = row.find('a', href=True).text

            # 提取链接
            link = f"https://comments.ustr.gov{row.find('a', href=True)['href']}"
        
            print(f"Submission ID: {submission_id}")
        
            # 打开链接网页并切换到新窗口
            driver.execute_script("window.open();")
            driver.switch_to.window(driver.window_handles[1])
            driver.get(link)
        
            # 添加睡眠时间
            time.sleep(3)
        
            try:
                # 等待一段时间以确保页面加载完成
                wait = WebDriverWait(driver, 5)
                wait.until(EC.presence_of_element_located(("xpath", "//a[@title='Download']")))

                # 设置下载文件夹路径
                download_dir = f"/Users/isabellawu/Desktop/RA - Bingjing -Web crawling/2023-0001/USTR-2023-0001/{submission_id}"  # 请替换为您希望保存文件的文件夹路径

                # 创建下载文件夹
                if not os.path.exists(download_dir):
                    os.makedirs(download_dir)

                # 查找下载链接
                download_links = driver.find_elements("xpath", "//a[@title='Download']")

                # 循环遍历下载链接并下载文件
                for index, download_link in enumerate(download_links, start=1):
                    download_url = download_link.get_attribute("href")
                    print(download_url)

                    # 发送HTTP请求并下载文件
                    response = requests.get(download_url)

                    content_type = response.headers.get('Content-Type', '')
                    print(content_type)

                    # 根据Content-Type确定文件扩展名
                    if 'pdf' in content_type:
                        file_extension = 'pdf'
                    elif 'xlsx' in content_type:
                        file_extension = 'xlsx'
                    elif 'png' in content_type:
                        file_extension = 'png'
                    else:
                    # 如果无法确定文件类型，默认使用未知扩展名(检查是否能打开)
                        file_extension = 'unknown'

                    # 使用文件类型创建文件名
                    filename = f"{submission_id}_{index}.{file_extension}"

                    # 构建完整的文件路径
                    file_path = os.path.join(download_dir, filename)

                    # 检查HTTP响应状态码
                    if response.status_code == 200:
                        with open(file_path, "wb") as file:
                            file.write(response.content)
                        print(f"下载成功: {file_path}")
                    else:
                        print(f"下载失败: {download_url}")

            except Exception as e:
                print(f"未找到下载链接")

            # 关闭链接网页并切换回原窗口
            driver.close()
            driver.switch_to.window(driver.window_handles[0])
 
    # 查找"Next"按钮
    next_button = driver.find_element("xpath", "//button[contains(text(), 'Next')]")
    
    if "aria-disabled=\"true\"" in next_button.get_attribute("outerHTML"):
        # 如果"Next"按钮不可用，退出循环
        break

    # 点击"Next"按钮
    next_button.click()

# 关闭WebDriver
driver.quit()

```
