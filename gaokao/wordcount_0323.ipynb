```python
import pandas as pd

addr = pd.read_excel("/Users/isabellawu/Dropbox/China Higher Education/data/major lists/招生计划/plan_data/address-985211.xlsx")
location_str = addr["地址"]
```

```python
import cpca
df = cpca.transform(location_str)
merged_df = pd.concat([addr, df], axis=1)
```

```python
def replace_city_with_province(row):
    if row['市'] == '市辖区':
        return row['省']
    else:
        return row['市']

# 使用apply方法应用自定义函数
merged_df['市'] = merged_df.apply(replace_city_with_province, axis=1)
```

```python
import pandas as pd

# 替换为你的 Excel 文件路径
file_path = '/Users/isabellawu/Desktop/RA-Bingjing/工作-报告-地级市.xlsx'

# 读取 Excel 文件
df1 = pd.read_excel(file_path)
df2 = pd.read_stata('/Users/isabellawu/Desktop/RA-Bingjing/省市代码.dta')
df = pd.merge(df1, df2, on='地区')
df = df.replace("\n", "", regex=True)

df.head(10)
```

```python
# 计算“新能源”和“新能源汽车”的出现次数
df['新能源'] = df['报告全文'].str.count('新能源')
df['新能源汽车'] = df['报告全文'].str.count('新能源汽车')
df['太阳能'] = df['报告全文'].str.count('太阳能')
df['光伏'] = df['报告全文'].str.count('光伏')
df['电池']= df['报告全文'].str.count('电池')
df['人工智能'] = df['报告全文'].str.count('人工智能')
df['AI'] = df['报告全文'].str.count('AI')
df['大数据']= df['报告全文'].str.count('大数据')
df['芯片']= df['报告全文'].str.count('芯片')
df['集成电路']= df['报告全文'].str.count('集成电路')
df['半导体']= df['报告全文'].str.count('半导体')

# 分组并计算每个省份每年的总计数
grouped_df = df.groupby(['年份', '地区', '省']).agg({
    '新能源' : 'sum',
    '新能源汽车' : 'sum',
    '太阳能' : 'sum',
    '光伏' : 'sum',
    '电池' : 'sum',
    '人工智能': 'sum',
    'AI': 'sum',
    '大数据': 'sum',
    '芯片': 'sum',
    '集成电路': 'sum',
    '半导体': 'sum',


}).reset_index()

grouped_df['solar'] = grouped_df['太阳能'] +grouped_df['光伏']
grouped_df['AI_人工智能'] = grouped_df['AI'] + grouped_df['人工智能'] + grouped_df['大数据'] 
grouped_df['chip'] = grouped_df['芯片'] + grouped_df['集成电路'] + grouped_df['半导体'] 
grouped_df['ev'] = grouped_df['太阳能'] +grouped_df['光伏']


grouped_df2 = grouped_df[grouped_df['年份']>=2014]

```

```python
# 指定输出 Excel 文件的路径
output_file_path = '/Users/isabellawu/Desktop/RA-Bingjing/wordcount.xlsx'

# 将数据写入 Excel 文件
grouped_df2.to_excel(output_file_path, index=False, engine='openpyxl')
```
